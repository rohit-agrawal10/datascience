```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
from IPython.display import display

matplotlib.rc("figure", figsize=(8,6))
matplotlib.rc("axes", labelsize=16, titlesize=16)
matplotlib.rc("xtick", labelsize=14)
matplotlib.rc("ytick", labelsize=14)
matplotlib.rc("legend", fontsize=14)
matplotlib.rc("font", size=14)

df1 = pd.read_excel('DDS9_Data_Extract_with_labels.xlsx')
df2 = pd.read_excel('DDS10_Data_Extract_with_labels.xlsx')
df3 = pd.read_excel('DDS11_Data_Extract_with_labels.xlsx')
```

## Nature of Subscription Services

```python
new_label = []
new_label.append("Pay TV")
new_label.append("Home internet")
new_label.append("Landline telephone")
new_label.append("Mobile voice")
new_label.append("Mobile data plan")
new_label.append("Streaming video service")
new_label.append("Streaming music service")
new_label.append("Gaming")
new_label.append("News")
new_label.append("Magazine")
new_label.append("None")
```

```python
for i in range(0, len(new_label)):
    df1[new_label[i]].replace(('Yes', 'No'), (1, 0), inplace=True)
    df2[new_label[i]].replace(('Yes', 'No'), (1, 0), inplace=True)
    df3[new_label[i]].replace(('Yes', 'No'), (1, 0), inplace=True)
```

```python
df_heatmap = df1[new_label]
corr = df_heatmap.corr()
plt.clf()
sns.heatmap(corr)
plt.show()
```

###### High Correlations between News and Magazine, Streaming Video and Streaming Audio Services

```python
subscription_2009= {}
subscription_2010= {}
subscription_2011= {}
for label in new_label:
    subscription_2009[label] = df1[label].sum()
    subscription_2010[label] = df2[label].sum()
    subscription_2011[label] = df3[label].sum()
```

```python
# data to plot
n_groups = len(new_label)
val_2009 = subscription_2009.values()
val_2010 = subscription_2010.values()
val_2011 = subscription_2011.values()
 
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8

rects1 = plt.bar(index, val_2009, bar_width,
                 alpha=opacity,
                 label='2009')

rects2 = plt.bar(index+ bar_width, val_2010, bar_width,
                 alpha=opacity,
                 label='2010')

rects3 = plt.bar(index+ (2*bar_width), val_2011, bar_width,
                 alpha=opacity,
                 label='2011')
 
 
plt.ylabel('Count of subscribers')
plt.xlabel('Subscription type')
plt.title(' Count of subscribers by year')
plt.xticks(index + (bar_width), subscription_2010.keys(), rotation = 90)
plt.legend()
 

plt.show()
```

##### Home Internet has the highest subscribers. Streaming Video and Audio subscriptions and Gaming are on the rise

## Binge Watching TV Series

```python
old_label = "QNEW24 - Do you ever \"binge-watch\" television shows, meaning watching three or more episodes of a TV series in one sitting?"
new_label = "Binge Watch"
df1.rename(columns={old_label : new_label}, inplace=True)
df2.rename(columns={old_label : new_label}, inplace=True)
df3.rename(columns={old_label : new_label}, inplace=True)
```

```python
df1[new_label].replace(('Yes', 'No'), (1, 0), inplace=True)
df2[new_label].replace(('Yes', 'No'), (1, 0), inplace=True)
df3[new_label].replace(('Yes', 'No'), (1, 0), inplace=True)
```

```python
# data to plot
bw_2010 = {}
bw_2011 = {}
bw_2012 = {}

n_groups = 1
bw_2010['Yes'] = float(100 * df1[new_label].sum()) / float(len(df1))
bw_2011['Yes'] = float(100 * df2[new_label].sum()) / float(len(df2))
bw_2012['Yes'] = float(100 * df3[new_label].sum()) / float(len(df3))

val_2010 = bw_2010.values()
val_2011 = bw_2011.values()
val_2012 = bw_2012.values()
 
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.05
opacity = 0.8
 
rects1 = plt.bar(index, val_2010, bar_width,
                 alpha=opacity,
                 label='2010')
 
rects2 = plt.bar(index + bar_width, val_2011, bar_width,
                 alpha=opacity,
                 label='2011')

rects3 = plt.bar(index + (2*bar_width), val_2012, bar_width,
                 alpha=opacity,
                 label='2012')
 
plt.ylabel('Percentage of Binge Watchers')
plt.title('Percentage of binge watchers by year')
plt.xticks(index + bar_width, bw_2010.keys())
plt.legend()
 

plt.show()

```

##### Hence, we can see we have around 65-70% responses who binge watch shows. This is significant
